## Tidiness Issues

#### Define
Contacts column in `Patients Table` need to be split into email and phone number

#### Code

```python
patients_clean['email']=patients_clean['contact'].str.extract(r'([A-Za-z]+@[a-z]+\.[a-z]+)')
```

```python
patients_clean['phone']=patients_clean['contact'].str.extract(r'([\+\d]*[\s(]*[\d]{3}[\s\)\-]+\d{3}[\s\)\-]\d{3})')
```

```python
patients_clean.drop(columns="contact",axis=1,inplace=True)
```

#### Test

```python
patients_clean.sample(10)
```

#### Define

Three variables in two columns in **`treatments_clean`** df (treatment, start dose and end dose).Will use the pandas melt function and str.split() to make auralin novodra into a column named `drug_type` and other columns  `start_dose` and `end_dose` 

#### Code:

```python
a=pd.melt(treatments_clean,id_vars=["given_name","surname","hba1c_start","hba1c_end","hba1c_change"],var_name="drug_type",value_name="Dosage")
```

```python
a["Dosage"]=   a["Dosage"].str.extract(r'([\d]{2}[u]\s[\-]\s[\d]{2}[u])')
```

```python
a.dropna(subset=["Dosage"],inplace=True)
```

```python
a["start_dose"],a["end_dose"]=a.Dosage.str.split("-",1).str
```

```python
a.drop(columns="Dosage",axis=1,inplace=True)
```

```python
treatments_clean=a.copy()
```

#### Test

```python
treatments_clean.sample(5)
```

#### Define

**`Adverse Reactions`** should be a part of **`Treatments`** Table. We will merge them using merge function joining on `given_name` and `surname`

#### Code

```python
treatments_clean=pd.merge(treatments_clean,adverse_reactions_clean,how="left", on=['given_name','surname'])
```

#### Test

#### Define:

Given name and surname columns in `patients` table duplicated in `treatments` and `adverse_reactions` tables.We will replace that column with patient_id

#### Code:

```python
b=patients_clean["surname"].to_frame()
c=patients_clean["given_name"].to_frame()
d=patients_clean["patient_id"].to_frame()
```

```python
pat_id=pd.concat([b,c,d],axis=1)
pat_id.given_name=pat_id.given_name.str.lower()
pat_id.surname=pat_id.surname.str.lower()
```

```python
treatments_clean=pd.merge(pat_id,treatments_clean,on=["given_name","surname"])
```

```python
treatments_clean.drop(["given_name","surname"],axis=1,inplace=True)
```

#### Define:
There are  abbreviations used sometimes and full state name used other times in the state column of `patients` table.Will convert the full state name into abbreviations.

#### Code

```python
state_abbrev = {'California': 'CA',
                'New York': 'NY',
                'Illinois': 'IL',
                'Florida': 'FL',
                'Nebraska': 'NE'}


def abbreviate_state(patient):
    if patient['state'] in state_abbrev.keys():
        abbrev = state_abbrev[patient['state']]
        return abbrev
    else:
        return patient['state']
    
patients_clean['state'] = patients_clean.apply(abbreviate_state, axis=1)
```
